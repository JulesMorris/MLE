```python
def chr2int(s: str) -> int:
    """Converts a letter to an integer.
    Offset for ascii included. E.g.,
    a=0, b=1, c=2, ..."""
    return ord(s.lower()) - ord("a")


def int2chr(i: int) -> str:
    """Converts an integer to a letter.
    Offset for ascii included. E.g.,
    0=a, 1=b, 2=c, ..."""
    return chr(i + ord("a"))


def sub(a: int, b: int) -> int:
    """Subtracts two integers mod 26. E.g.,
    negative values will wrap around back to the 
    end af the alphabet."""
    return (a - b) % 26


ciphertext = "Ahmvvrchmodiemsakizoneofztallr"

codes = [
    "booz",
    "allen",
    "hamilton",
    "national",
    "business",
    "park",
    "turing",
    "brutus",
    "salad",
    "consultant",
]

for code in codes:
    codeword = (code * 20)[: len(ciphertext)]
    print(ciphertext)
    print(codeword)
    for _1, _2 in zip(codeword, ciphertext):
        print(int2chr(sub(chr2int(_2), chr2int(_1))), end="")
    print("\n")
```
