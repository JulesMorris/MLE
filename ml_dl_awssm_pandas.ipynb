### Subsetting
- Similar to SQL statement select * where (condition)...

```python
#show last 10 rows where Pclass equals 1
df[df.Pclass == 1].tail(10)
```

```python
#get first five entries of first class passengers that are also male
df[(df.Pclass ==1) & (df.Sex == 'male')].head() 
```

```python
# get entries for those that are either first class passengers or male (can get first class and male using this,
# but will return if either condition is true)
df[(df.Pclass == 1) | (df.Sex == 'male')]
```

```python
cond1 = df.Pclass > 1
cond2 = df.Sex == 'male'

df[cond1 & cond2].head()
```

## Map and Apply
- More efficient to use `map` or `apply` than to use functions one by one

```python
df.sort_index(inplace = True)
```

```python
# we want to grab the last name
df.Name.values[-10:][0]
```

```python
#split at the comma and not get the 0 index item (the family/first item name)
"Markun, Mr. Johann".split(",")
```

```python
#split at the comma
"Markun, Mr. Johann".split(",")[0]
```

```python
def get_first_name(x):
    return x.split(',')[0]
```

```python
#create a column for Last_name
df['Last_name'] = df['Name'].map(lambda x: x.split(',')[0])
df.sample(10)
```

```python
#want to combine name and age in a single sentence

df.apply(lambda x: x['Name'] + " "  + str(x['Age']), axis = 1) #apply goes through columns, axis = 1 row-wise function
```

```python
#get the log of each fare
#import numpy as np
df['log_fare'] = df.Fare.map(lambda x:np.log(x + 1)) #must add plus one because there were free tickets that cost 0 monies
df.sample(10)
```

```python
def log_fn(x):
    return np.log(x + 1)

df['log_fare2'] = df.Fare.map(log_fn)
```

```python
df.sample(10)
```

# Groupby

```python
#(what you want to group by)[what you want to aggregate].aggregation()
df.groupby("Pclass")['log_fare'].mean()
```

#### How many nulls are there in Age?

```python
df['Age'].isnull().sum()
```

```python
#how to use apply to get the number of nulls per class
df.groupby("Pclass")['Age'].apply(lambda x:x.isnull().sum())
```

```python
#pass in a dictionary to use .agg, and then use a list of desired aggregation functiobs
df.groupby('Pclass').agg({'Age': [sum, min, max, np.mean]})
```
